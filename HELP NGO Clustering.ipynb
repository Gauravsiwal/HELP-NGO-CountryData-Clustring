### Preprocessing

```python
# Check for missing values
data.isnull().sum().sum()

# No Missing Values
```

```python
# Lets treat the extreme outliers
# We will go with capping using IQR method
# In order to cap only the extreme ones lets keep whisker lenghth = 2.5

for i in data.columns:
    q1,q3 = np.quantile(data[i],[0.25,0.75])
    iqr = q3-q1
    ul,ll = q3+2.5*iqr,q1-2.5*iqr
    data[i] = data[i].apply(lambda x: ul if x>ul else ll if x<ll else x)
```

```python
t=1
plt.figure(figsize=(15,10))
for i in data.columns:
    plt.subplot(3,3,t)
    sns.boxplot(data=data,x=i,color='green')
    plt.title(f'Skewness:{round(data[i].skew(),2)}')
    t+=1
plt.tight_layout()
plt.show()
```

```python
# Lets apply PCA on the data before we procced with clustring. Just have better visualization
# and better clsutring.
# In pca we need to pass standardized data. So lets import the functions from libraries


# Lets create and instance of scalling
scaler = StandardScaler()

# Also lets create an instance for PCA
pca = PCA(n_components=5,random_state=2)

# Lets create pipeline for the above two proccess
pipeline = Pipeline([('Scale',scaler),
                     ('PCA',pca)])
```

```python
pipeline.fit(data)
```

```python
pca.explained_variance_ratio_.cumsum()

# At n_components = 5 we are able to explain ~ 95% variance , which is good enough.
# so 9-Dimension data can be reduced to 5-Dimension
```

```python
final_data = pipeline.transform(data)
```

```python
# .transform return the array. Lets convert it into dataframe
final_data = pd.DataFrame(final_data,columns=['PC1','PC2','PC3','PC4','PC5'],
                          index=data.index)
final_data.head(3)
```

## K-Means Clustering

```python
# Identify the value of k

# KElbow Visualizer
kelbow = KElbowVisualizer(estimator=KMeans(random_state=2),k=10)
```

```python
kelbow.fit(final_data)
kelbow.show()
```

Although the above plot marks the elbow at k=6. But we can see the significant drop in distortion score from k=2 to k=3. As we increase the number of cluster it might become over fitted. Lets explore sillhouette visualizer to identify the correct or more generalized value of k.

```python
t = 1
plt.figure(figsize=(15,10))
for i in range(2,11):
    plt.subplot(3,3,t)
    model = KMeans(n_clusters=i,random_state=2)
    sil_viz = SilhouetteVisualizer(model)
    sil_viz.fit(final_data)
    plt.title(f'K:{i} Sillhouette Score: {round(sil_viz.silhouette_score_,4)}')
    t+=1
plt.tight_layout()
plt.show()
```

After analyzing Elbow plot and Sillhoette visulizer lets go with k=3. Although there are few datapoints that have been misclassified, but we see no drop in sillhouette score and there is good enough data in every cluster. Also from k-elbow plot significant drop in WCSS (Distortion score) was observed from k=2 to k=3.

```python
final_kmeans = KMeans(n_clusters=3,random_state=2)
```

```python
labels = final_kmeans.fit_predict(final_data)
```

```python
# Lets check the sillhouette score
print('Silhouette Score:',silhouette_score(final_data,labels))
```

```python
# Lets put these lables on PCA data
final_data['Labels'] = labels
final_data.head(3)
```

```python
# Lets Visulize the clsuters
# In order get the best visualization lets choose PC1 and PC2 as they have maximum variance
centroids = final_kmeans.cluster_centers_
centroids

sns.scatterplot(data=final_data,x='PC1',y='PC2',palette=['m','b','green'],hue='Labels')
for i in range(3):
    plt.plot(centroids[i][0],centroids[i][1],marker='*',ms=12,color='k')
plt.show()
```
